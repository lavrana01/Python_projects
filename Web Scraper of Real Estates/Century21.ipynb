```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.century21.com/real-estate/rock-springs-az/LCAZROCKSPRINGS/")
c = r.content
c
```

```python
soup = BeautifulSoup(c, 'html.parser')
```

```python
all = soup.find_all('div',{'class': 'property-card-primary-info'})
```

```python
all[0].find_all('a',{'class': 'listing-price'})[0].text.replace('\n','').replace(' ','')
```

```python
l = []
for item in all:
    d = {}
    d['price'] = item.find_all('a',{'class': 'listing-price'})[0].text.replace('\n','').replace(' ','')
    d['Address'] = item.find_all('div',{'class': 'property-address-info'})[0].text.replace('\n','')
    
    try:
        d['beds'] = item.find('div',{'class': 'property-beds'}).text.replace('\n','')
    except:
        d['beds'] = None
    try:
        d['area'] = item.find('div',{'class': 'property-sqft'}).text.replace('\n','')
    except:
        d['area'] = None
    try:
        d['baths'] = item.find('div',{'class': 'property-baths'}).text.replace('\n','')
    except:
        
        d['baths'] = None
    
    l.append(d)
l
```

```python
import pandas
df = pandas.DataFrame(l)
df
```

```python
df.to_csv("Output.csv")
```
